```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
resources_path = '../lab2/src/test/resources/'

def data_path(func_name):
    return "{}{}_example.csv".format(resources_path, func_name)

sin_data = pd.read_csv(data_path('sin'))
cos_data = pd.read_csv(data_path('cos'))
tan_data = pd.read_csv(data_path('tan'))
cot_data = pd.read_csv(data_path('cot'))

tan_data = tan_data[abs(tan_data['Function']) <= 10]
cot_data = cot_data[abs(cot_data['Function']) <= 10]

ln_e_data = pd.read_csv(data_path('ln'))
log2_data = pd.read_csv(data_path('log2'))
log3_data = pd.read_csv(data_path('log3'))
log5_data = pd.read_csv(data_path('log5'))

func_data = pd.read_csv(data_path('func'))
func_data = func_data[abs(func_data['Function']) <= 10]
```

```python
def orig_function(x): 
    if x <= 0:
        return (((((np.cos(x) ** 2) / np.tan(x)) + np.sin(x)) * (np.cos(x) * np.cos(x))) - (1/np.tan(x) + np.tan(x)))
    else:
        return (((((np.log(x)/np.log(3) - np.log2(x)) ** 2) ** 2) ** 3) / (np.log(x)/np.log(5)))

def prepare_graph(data, name, size = (8, 5)):
    plt.figure(figsize=size)
    plt.grid(True)
    plt.axhline(y=0, color='black')
    plt.axvline(x=0, color='black')
    plt.xlabel("X")
    plt.ylabel("Function(X)=Y")
    plt.plot(data['X'], data['Function'], label=name)
    plt.legend()
```

```python
prepare_graph(sin_data, 'Sin(X)')
plt.show()

prepare_graph(cos_data, 'Cos(X)')
plt.show()

prepare_graph(tan_data, 'Tan(X)')
plt.show()

prepare_graph(cot_data, 'Cot(X)')
plt.show()

prepare_graph(ln_e_data, 'Ln(X)')
plt.show()

prepare_graph(log2_data, 'Log_2(X)')
plt.show()

prepare_graph(log3_data, 'Log_3(X)')
plt.show()

prepare_graph(log5_data, 'Log_5(X)')
plt.show()
```

```python
prepare_graph(func_data, 'General_Func(X)', size = (16, 10))
plt.scatter(func_data['X'][::5], [orig_function(x) for x in func_data['X'][::5]], 
            label="Original_Func(X)", c = 'red', s = 15, alpha = 0.25)
plt.show()
```
